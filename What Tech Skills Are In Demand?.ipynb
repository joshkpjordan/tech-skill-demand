# What Tech Skills Are In Demand?

We scraped 2310 job posts for IT and technology jobs from one of NZ's top job websites (seek.co.nz).
The values below represent the counts for the mentions of each word or phrase. We used NLTK tokenize, ngram and counter libraries to establish the count.

```python
import pandas as pd
from matplotlib import pyplot as plt

ss_lab = ["SSIS","SSRS","SSAS"]
ss_vals = [246,233,246]

cloud_lab = ["Azure","AWS","GCP"]
cloud_vals = [1045, 809, 63]

cisco_lab = ["CCDP","CCSP","CCNP","CNNA"]
cisco_vals = [9,6,0,6]

analytics_lab = ["Power Bi", "Tableau", "Alteryx", "Hadoop"]
analytics_vals = [30,61,6,39]

methods_lab = ["Scrum","Lean","Waterfall","Agile","Kaizen","Devops","six sigma"]
methods_vals = [349,186,72,2041,72,864,10]

languages_lab = ["Ruby","Java", "Python", "Clojure", "Kotlin", "C++", "Javascript", "Rust", "Go", "Perl","PHP","Ajax","Matlab","SAS","SQL", "Scala"]
languages_vals = [102,484,279,3,66,595,789,3,9,24,294,6,12,57,792,45]

ml_lab = ["Tensorflow", "Pytorch"]
ml_vals = [9,6]

jsframework_lab = ["Vue", "React"]
jsframework_vals = [93,569]

backendframework_lab = ["Rails", "Laravel", "Node"]
backendframework_vals = [72,66,140]

developer_lab = ["Frontend", "Backend", "Fullstack"]
developer_vals = [513,311,270]

os_lab = ["Android", "IOS"]
os_vals = [234,153]

platform_lab = ["Shopify", "Wordpress"]
platform_vals = [42,33]

s_lab = ["Statistics", "Machine Learning"]
s_vals = [169, 110]

softskills_lab = ["Communication Skills", "Problem Solving Skills", "Positive Attitude", "Work Ethic", "Stress Management Skills", "Analytical Skills"]
softskills_vals = [2365,1289,1018,522,654,429]
```

```python
plt.bar(range(len(developer_lab)), developer_vals)
ax=plt.subplot()
ax.set_xticks(range(len(developer_lab)))
ax.set_xticklabels(developer_lab)
plt.title("Frontend Vs. Backend Vs. Fullstack Developers")
plt.ylabel("Counts")
plt.xlabel("Developers")
plt.show()
```

There are more jobs for Frontend developers than Backend or Fullstack developers. It seems that employers prefer someone who specialises in either Frontend or Backend, than Fullstack (the jack of all trades).

```python
plt.bar(range(len(backendframework_lab)), backendframework_vals)
ax=plt.subplot()
ax.set_xticks(range(len(backendframework_lab)))
ax.set_xticklabels(backendframework_lab)
plt.title("Backend Framework Knowledge")
plt.ylabel("Counts")
plt.xlabel("Back-End Frameworks")
plt.show()
```

Surprisingly, the demand for Laravel is still strong, roughly equal with Ruby on Rails. If we were to go online, we'd see a lot of people talking about the demise of PHP (and it's frameworks by extension). But if we look at this graph, we can see that PHP and Laravel are still good options
But overall, Node, which is a Javascript framework has the most demand by a long shot and should be what you consider.

```python
plt.bar(range(len(jsframework_lab)), jsframework_vals)
ax=plt.subplot()
ax.set_xticks(range(len(jsframework_lab)))
ax.set_xticklabels(jsframework_lab)
plt.title("Vue vs. React")
plt.ylabel("Counts")
plt.xlabel("Front-End Framework")
plt.show()
```

Vue and React are both Javascrip frontend frameworks. As we can see, React is in much more demand. Looking at our graph above for frontend vs. backend developers. It makes sense to learn React if you were to learn frontend development.

```python
plt.bar(range(len(methods_lab)), methods_vals)
ax=plt.subplot()
ax.set_xticks(range(len(methods_lab)))
ax.set_xticklabels(methods_lab)
plt.title("Demand For Methodologies")
plt.ylabel("Counts")
plt.xlabel("Methodology")
plt.show()
```

Agile and Devops are the most popular methodologies by a long shot. Some employers prefer the other methodologies. The reason why needs further investigation.

```python
plt.figure(figsize=(15,8))
plt.bar(range(len(languages_lab)), languages_vals)
ax=plt.subplot()
ax.set_xticks(range(len(languages_lab)))
ax.set_xticklabels(languages_lab)
plt.title("Demand For Programming Language Knowledge")
plt.ylabel("Counts")
plt.xlabel("Programming Language")
plt.show()
```

SQL and Javascript are the most in-demand languages. This is followed by C++, Java, Python and PHP. It is interesting to see the level of continuing demand for PHP, which is roughly equal with Python. There is much discussion about the death of PHP, however strong demand still exists. The massive demand for Javascript reflects the popularity of javascript based frameworks Node and React. There are clear, niche markets for SAS, Ruby and Kotlin programming languages.

```python
plt.bar(range(len(ss_lab)),ss_vals)
ax = plt.subplot()
ax.set_xticks(range(0, len(ss_lab)))
ax.set_xticklabels(ss_lab)
plt.title("Demand For SQL Server Knowledge")
plt.ylabel("Counts")
plt.xlabel("MS Server Products")
plt.show()
```

Many jobs require knowledge of Microsoft Server. Here are three products with a decent amount of demand.

```python
plt.bar(range(len(cloud_lab)), cloud_vals)
ax=plt.subplot()
ax.set_xticks(range(len(cloud_lab)))
ax.set_xticklabels(cloud_lab)
plt.title("Demand For Cloud Service Knowledge")
plt.ylabel("Counts")
plt.xlabel("Cloud Service")
plt.show()
```

Interestingly, Azure beats out AWS when it comes to cloud services. This is unlikely to be the case in markets overseas where AWS is very popular. A common theme of here is the dominance of Microsoft products in the corporate environment. Bottom line is, get good at Microsoft products if you want a job.

```python
plt.bar(range(len(cisco_lab)), cisco_vals)
ax=plt.subplot()
ax.set_xticks(range(len(cisco_lab)))
ax.set_xticklabels(cisco_lab)
plt.title("Demand For Cisco Qualifications")
plt.ylabel("Counts")
plt.xlabel("CISCO Qualification")
plt.show()
```

Cisco qualifications are in demand. Specifically the CCDP qualification.

```python
plt.bar(range(len(analytics_lab)), analytics_vals)
ax=plt.subplot()
ax.set_xticks(range(len(analytics_lab)))
ax.set_xticklabels(analytics_lab)
plt.title("Demand For Analytics Software Knowledge")
plt.ylabel("Counts")
plt.xlabel("Analytical Software")
plt.show()
```

Tableau has more demand than Power BI (a microsoft product) when it comes to analytics software.

```python
plt.bar(range(len(ml_lab)), ml_vals)
ax=plt.subplot()
ax.set_xticks(range(len(ml_lab)))
ax.set_xticklabels(ml_lab)
plt.title("Pytorch vs. Tensorflow")
plt.ylabel("Counts")
plt.xlabel("Library")
plt.show()
```

Although, there aren't a huge number of Machine Learning jobs. At this point in time, there is more demand for Tensorflow library knowledge.

```python
plt.bar(range(len(os_lab)), os_vals)
ax=plt.subplot()
ax.set_xticks(range(len(os_lab)))
ax.set_xticklabels(os_lab)
plt.title("Demand For IOS and Android Developers")
plt.ylabel("Counts")
plt.xlabel("Operating System")
plt.show()
```

There is a greater demand for Android developers than Ios. This makes sense as the Android market is a lot bigger than the Ios market.

```python
plt.bar(range(len(platform_lab)), platform_vals)
ax=plt.subplot()
ax.set_xticks(range(len(platform_lab)))
ax.set_xticklabels(platform_lab)
plt.title("Demand For Wordpress and Shopify Developers")
plt.ylabel("Counts")
plt.xlabel("Platforms")
plt.show()
```

Right now, there is a greater demand for Shopify developers than Wordpress developers. This could be a distortion due to the Covid Pandemic (as of November 2020) where a lot of brick and mortar businesses are going online, and would use Shopify to do so.

```python
plt.bar(range(len(s_lab)), s_vals)
ax=plt.subplot()
ax.set_xticks(range(len(s_lab)))
ax.set_xticklabels(s_lab)
plt.title("Demand For Statistics vs. Machine Learning")
plt.ylabel("Counts")
plt.xlabel("Skills")
plt.show()
```

Despite the hype for Machine Learning, the amount of jobs is not quite there yet. 

```python
plt.figure(figsize=(15,8))
plt.bar(range(len(softskills_lab)), softskills_vals)
ax=plt.subplot()
ax.set_xticks(range(len(softskills_lab)))
ax.set_xticklabels(softskills_lab)
plt.title("Demand For Soft Skills By Employers in New Zealand")
plt.ylabel("Counts")
plt.xlabel("Soft Skills")
plt.show()
```

Communication skills are by far the most important skills requested by employers. This is almost double the next most requested skill, problem solving. If you want to get hired you should focus on soft skills, especially communication, problem solving while a positive attitude goes a long way as well.
The bottom line, people skills are important if you want to succeed.

# Limitations
Trying to understand the sentiment contained in words is by no means perfect and 100% inclusive.
This study applies to the New Zealand job market. Larger job markets will have a larger range of skills.
